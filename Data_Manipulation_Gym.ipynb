#Data Manipulation

##Grup

###Rata-rata waktu gym per tipe abonoment

```python
avg_time = df.groupby('abonoment_type')['avg_time_in_gym'].mean().reset_index()
avg_time.head()
```

###Hitung jumlah member per tipe membership

```python
df_member = df['abonoment_type'].value_counts().reset_index()
df_member.columns = ['abonoment_type', 'number_of_members']
df_member.head()
```

##Sort

Urutkan berdasarkan pengeluaran tertinggi (Monetary)

```python
top_customer = df.sort_values('monetary', ascending=False).head(10)
top_customer.head()
```

##Agregasi

Total pendapatan berdasarkan tipe abonoment

```python
income_per_type = df.groupby('abonoment_type')['monetary'].sum().reset_index()
income_per_type.head()
```

###Tambah kolom baru

Kategori umur

```python
def age_category(age):
    if age < 20:
        return 'Teenager'
    elif age < 35:
        return 'Young Adult'
    elif age < 50:
        return 'Mature'
    else:
        return 'Elderly'

df['age_category'] = df['age'].apply(age_category)
```

###Gabung data

Join antara df dan member

```python
df_merged = pd.merge(df, df_member, on='abonoment_type', how='left')
```
