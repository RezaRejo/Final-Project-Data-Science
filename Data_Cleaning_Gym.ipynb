#Data Cleaning

```python
df.info()
df.describe(include='all')
```

##Cek Duplicat

```python
#cek ada duplicat atau tidak
df.duplicated().sum()
```

```python
#Menstandarkan nama kolom
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

##Cek null

```python
print(df.isnull().sum())
```

Kita drop nilai null dan tidak perlu digunakan

```python
df_clean = df.drop(columns=['fav_group_lesson', 'fav_drink', 'name_personal_trainer'])
df_clean.head()
```

##Format tanggal

```python
if 'birthday' in df.columns:
    df['birthday'] = pd.to_datetime(df['birthday'], errors='coerce')
if 'last_visit_date' in df.columns:
    df['last_visit_date'] = pd.to_datetime(df['last_visit_date'], errors='coerce')
```

Memastikan konsistensi kategori

```python
if 'gender' in df.columns:
    df['gender'] = df['gender'].str.strip().str.upper()
```

##Konversi tipe data numerik

```python
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = df[col].str.replace(',', '').astype(float)
        except:
            continue
```

##Groping gender dan abonoment type

```python
df_grouped = df.groupby(['gender', 'abonoment_type']).size().reset_index(name='count')
df_grouped
```
